```python
import pickle
import os

res_dir = "results"
def load_result(model_name,plain_mode = False):
  file_Name = os.path.join(res_dir,model_name+".pkl")
  pkl_file = open(file_Name, 'rb')
  f = pickle.load(pkl_file)
  test_acc = f['test_acc'] 
  train_acc = f['train_acc'] 
  best_acc = f['best_acc'] 
  pkl_file.close()
  return train_acc,test_acc, best_acc
```

### Loading the results of Resisual Network

```python
res32_train_acc,res32_test_acc, res32_best_acc = load_result("resnet32")
res20_train_acc,res20_test_acc, res20_best_acc = load_result("resnet20")
res44_train_acc,res44_test_acc, res44_best_acc = load_result("resnet44")
```

### Loading the results of plain Network

```python
# res32_train_acc,res32_test_acc, res32_best_acc = load_result("resnet32")
plain20_train_acc,plain20_test_acc, plain20_best_acc = load_result("plain20")
plain44_train_acc,plain44_test_acc, plain44_best_acc = load_result("plain44")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6), dpi=80)
X= range(1,len(res32_test_acc))
res32_err = [1- a/100 for a in res32_test_acc]
res20_err = [100- a for a in res20_test_acc]
res44_err = [100- a for a in res44_test_acc]

# plt.plot(X, res32_err[1:], color="blue", linewidth=2.5, linestyle="-", label ="res32")
plt.plot(X, res20_err[1:], color="red",  linewidth=2.5, linestyle="-", label = "res20")
plt.plot(X, res44_err[1:], color="green",  linewidth=2.5, linestyle="-", label ="res44")

plt.title("ResNet: test error on CIFAR10, the deeper network has less error")
plt.legend()
plt.xlabel("iter")
plt.ylabel("test error (%)")
plt.show()
```
